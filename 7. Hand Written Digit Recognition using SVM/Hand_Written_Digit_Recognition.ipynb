## Splitting into input(x) and output(y)
#### **.values** function used here to convert pandas dataframes to numpy arrays

```
x = dataset.iloc[:,1:].values
y = dataset.iloc[:,0].values
```

## Importing testing dataset

```
dataset_test = pd.read_csv("/content/drive/My Drive/Kaggle/Hand-Written-Digit-Recognition/test.csv")
```

```
dataset_test.iloc[:5,:]
```

```
x_test = dataset.iloc[:,1:].values
y_test = dataset.iloc[:,0].values
```

## Feature Scaling

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc = sc.fit(x)
x = sc.transform(x)
x_test = sc.transform(x_test)
```

```
x.shape
```

```
x[10000:10005,:]   # Values are normalised, although can't be seen here because of limited output
```

## Fitting the SVM Model

#### Import Library

```
from sklearn.svm import SVC
```

#### Building Classifier
##### **C** is regularisation parameter, where regularisation is inversely proportional to the value of C.
##### **Gamma** is a kernel parameter. Higher **gamma** value, higher chances of overfitting(less regularisation).
##### **Kernel rbf** maps values to higher dimensions or gaussian surfaces.

```
classifier = SVC(C=10,gamma='auto',kernel='rbf')
```

#### Fitting our model

```
classifier.fit(x,y)
```

## Predicting Test Set results

```
y_pred = classifier.predict(x_test)
```

## Accuracy

```
from sklearn.metrics.classification import accuracy_score
accu = accuracy_score(y_test,y_pred)
```

```
accu
```

## Confusion Matrix

```
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```
