```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("car_plate.jpg")
print(type(img))
```

```python
plt.imshow(img)
```

```python
no_plate = cv2.CascadeClassifier("haarcascade_russian_plate_number.xml")

plate = no_plate.detectMultiScale(img,1.1,10)

for (x,y,w,h) in plate:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),5)

plt.imshow(img)
```

## Loading the image again, applying blur this time

```python
def display(img):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ax.imshow(new_img)
```

```python
img = cv2.imread("car_plate.jpg")
print(type(img))
```

```python
display(img)
```

```python
plate = no_plate.detectMultiScale(img,1.1,10)

for (x,y,w,h) in plate:
    
    partial_img = img[y:y+h,x:x+w].copy()
    blurr_img = cv2.medianBlur(partial_img,7)
    img[y:y+h,x:x+w] = blurr_img
```

```python
display(img)
```
